```python
a = QuantumRegister(33, 'q')
multiplexer = QuantumCircuit(a, name='Multiplexer')

#column 1
#minterms
m1=6
m2=7
m3=8
#controls
c1=4
c2=5
#Ternary Shannon + Equation 5
#11,10,01,00
for x in range(0, 9):
    #ternary shannon-->11
    multiplexer.x(c2)
    multiplexer.mct([c1,c2,m1],m2) #Eq 5a-->10
    multiplexer.mct([c1,c2,m1],m3) #Eq 5a
    multiplexer.x(c1)
    multiplexer.x(c2)
    multiplexer.mct([c1,c2,m2],m1) #Eq 5b-->01
    multiplexer.mct([c1,c2,m2],m3) #Eq 5b
    multiplexer.x(c1)
    multiplexer.mct([c1,c2,m3],m1) #Eq 5c-->00
    multiplexer.mct([c1,c2,m3],m2) #Eq 5c
    multiplexer.x(c1)
    multiplexer.x(c2)
    #multiplexer.barrier()
    m1+=3
    m2+=3
    m3+=3
    
#column 2

m1-=27
m2-=25
m3-=23 

c1-=2
c2-=2
for i in range(0, 3):
    for j in range(0, 3):
        #ternary shannon-->11
        multiplexer.x(c2)
        multiplexer.mct([c1,c2,m1],m2) #Eq 5a-->10
        multiplexer.mct([c1,c2,m1],m3) #Eq 5a
        multiplexer.x(c1)
        multiplexer.x(c2)
        multiplexer.mct([c1,c2,m2],m1) #Eq 5b-->01
        multiplexer.mct([c1,c2,m2],m3) #Eq 5b
        multiplexer.x(c1)
        multiplexer.mct([c1,c2,m3],m1) #Eq 5c-->00
        multiplexer.mct([c1,c2,m3],m2) #Eq 5c
        multiplexer.x(c1)
        multiplexer.x(c2)
        m1+=1
        m2+=1
        m3+=1
    m1+=6
    m2+=6
    m3+=6
#column 3
m1-=27
m2-=21
m3-=15 

c1-=2
c2-=2
for x in range(0, 9):
    #ternary shannon-->11
    multiplexer.x(c2)
    multiplexer.mct([c1,c2,m1],m2) #Eq 5a-->10
    multiplexer.mct([c1,c2,m1],m3) #Eq 5a
    multiplexer.x(c1)
    multiplexer.x(c2)
    multiplexer.mct([c1,c2,m2],m1) #Eq 5b-->01
    multiplexer.mct([c1,c2,m2],m3) #Eq 5b
    multiplexer.x(c1)
    multiplexer.mct([c1,c2,m3],m1) #Eq 5c-->00
    multiplexer.mct([c1,c2,m3],m2) #Eq 5c
    multiplexer.x(c1)
    multiplexer.x(c2)
    m1+=1
    m2+=1
    m3+=1
multiplexer.to_gate()
#multiplexer.draw('mpl')
```

```python
#Mirror for multiplexer
mirrormultiplexer = QuantumCircuit(33, name='Mirror Multiplexer')

#minterms
#m1=18
#m2=27
#m3=36
m1=15
m2=24
m3=33
#controls
c1=0
c2=1
#column 3
for x in range(0, 9):
    m1-=1
    m2-=1
    m3-=1
    mirrormultiplexer.x(c1)
    mirrormultiplexer.x(c2)
    mirrormultiplexer.mct([c1,c2,m3],m2) #Eq 5c
    mirrormultiplexer.mct([c1,c2,m3],m1) #Eq 5c-->00
    mirrormultiplexer.x(c1)
    mirrormultiplexer.mct([c1,c2,m2],m3) #Eq 5b
    mirrormultiplexer.mct([c1,c2,m2],m1) #Eq 5b-->01
    mirrormultiplexer.x(c2)
    mirrormultiplexer.x(c1)
    mirrormultiplexer.mct([c1,c2,m1],m3) #Eq 5a   
    mirrormultiplexer.mct([c1,c2,m1],m2) #Eq 5a-->10
    mirrormultiplexer.x(c2)
    #ternary shannon-->11
m1+=27
m2+=21
m3+=15 

c1+=2
c2+=2

for i in range(0, 3):
    m1-=6
    m2-=6
    m3-=6
    for j in range(0, 3):
        m1-=1
        m2-=1
        m3-=1
        mirrormultiplexer.x(c1)
        mirrormultiplexer.x(c2)
        mirrormultiplexer.mct([c1,c2,m3],m2) #Eq 5c
        mirrormultiplexer.mct([c1,c2,m3],m1) #Eq 5c-->00
        mirrormultiplexer.x(c1)
        mirrormultiplexer.mct([c1,c2,m2],m3) #Eq 5b
        mirrormultiplexer.mct([c1,c2,m2],m1) #Eq 5b-->01
        mirrormultiplexer.x(c2)
        mirrormultiplexer.x(c1)
        mirrormultiplexer.mct([c1,c2,m1],m3) #Eq 5a   
        mirrormultiplexer.mct([c1,c2,m1],m2) #Eq 5a-->10
        mirrormultiplexer.x(c2)
        #ternary shannon-->11
m1+=27
m2+=25
m3+=23 

c1+=2
c2+=2
for x in range(0, 9):
    m1-=3
    m2-=3
    m3-=3
    mirrormultiplexer.x(c1)
    mirrormultiplexer.x(c2)
    mirrormultiplexer.mct([c1,c2,m3],m2) #Eq 5c
    mirrormultiplexer.mct([c1,c2,m3],m1) #Eq 5c-->00
    mirrormultiplexer.x(c1)
    mirrormultiplexer.mct([c1,c2,m2],m3) #Eq 5b
    mirrormultiplexer.mct([c1,c2,m2],m1) #Eq 5b-->01
    mirrormultiplexer.x(c2)
    mirrormultiplexer.x(c1)
    mirrormultiplexer.mct([c1,c2,m1],m3) #Eq 5a   
    mirrormultiplexer.mct([c1,c2,m1],m2) #Eq 5a-->10
    mirrormultiplexer.x(c2)
    #ternary shannon-->11

mirrormultiplexer.to_gate()
#mirrormultiplexer.draw('mpl')
```
